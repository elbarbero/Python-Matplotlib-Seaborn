<h1>EJERCICIO</h1>
<p>Visualizar un gráfico de Esperanza de Vida frente a Renta per capita
¿Existe una correlación entre estas variables?</p>

```python
df_pais = pd.read_csv(r"E:\Python – Visualización Matplotlib & Seaborn\Python-Matplotlib-Seaborn\Datasets+Data+Science+con+Python+-+Visualización\1.3\Info_pais.csv", encoding = "ISO-8859-1", delimiter = ';')
```

```python
df_order = df_pais.sort_values("Esperanza de vida", ascending = True)
```

```python
df_order.head()
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
```

```python
df_order["Pobl_norm"] = df_order["Poblacion"]/(max(df_order["Poblacion"])/10000)
# se divide por 10000 para que la cantidad no salga tan pequeña. Es decir, que ahora la relación será 1:10000
```

```python
df_order
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)
```

```python
# nos muestra un color distinto en funcion de la poblacion normalizada
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"], c = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"], c = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)

# por ejemplo, cogemos los 10 primeros paises
for i in range(1, 10):
    plt.annotate(df_order["País"][i], (df_order["Renta per capita"][i], df_order["Esperanza de vida"][i]))
```

```python
total_rows = len(df_order.index)
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"], c = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)

for i in range(1, total_rows):
    plt.annotate(df_order["País"][i], (df_order["Renta per capita"][i], df_order["Esperanza de vida"][i]))
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"], c = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)

# print(list(df_order.iterrows())[1][1]["País"])
for row in df_order.iterrows():
    plt.annotate(row[1]["País"], (row[1]["Renta per capita"], row[1]["Esperanza de vida"]))
```

```python
plt.scatter(df_order["Renta per capita"], df_order["Esperanza de vida"], s = df_order["Pobl_norm"], c = df_order["Pobl_norm"])
plt.title("Renta per capita vs Esperanza de vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

# para hacer más grande el grafico
fig = plt.gcf()
fig.set_size_inches(14.5, 10)

# por ejemplo, cogemos los 10 primeros paises
for i in range(1, 10):
    plt.annotate(df_order["País"][i], (df_order["Renta per capita"][i], df_order["Esperanza de vida"][i]))

plt.yticks(ticks = range(1, 120, 10))
```

```python
from scipy import stats
```

```python
pendiente, intercepto, r_value, p_value, std_err = stats.linregress(df_ext["Periodo num_int"], df_ext["Total Activos"])
```

```python
sns.regplot(data=df_ext, x="Periodo num_int", y="Total Activos", scatter=True, order=1, color='green', label='order', line_kws={'label' : "y={0:.1f}x+{1:.1f}".format(pendiente, intercepto)})
plt.legend()
```

<p>Cargar el fichero xlsx de tráfico peatonal por distrito. A partir de gráfico violinplot trate de
dar respuesta a estas preguntas:</br>
- ¿Qué distrito tiene un rango dinámico de número de peatones más elevado? (diferencia entre el
máximo y mínimo valor)</br>
- Respecto a la función densidad probabilidad el distrito Chamberí, ¿es una distribución gaussiana?
En caso negativo, ¿podría dar respuesta a por qué no lo es? (Pista: Realice un gráfico a partir de
una sentencia groupby por cada hora)</br>
</p>

```python
df_peatones = pd.read_csv(r"E:\Python – Visualización Matplotlib & Seaborn\Python-Matplotlib-Seaborn\Datasets+Data+Science+con+Python+-+Visualización\3.5\PEATONES_2020_mod.csv", encoding = "ISO-8859-1", delimiter = ';')
df_peatones.head()
```

```python
df_grouping = df_peatones.groupby(["DISTRITO"])["PEATONES"].agg(["min", "max"])
df_grouping["Diff."] = df_grouping["max"] - df_grouping["min"]
resp = pd.DataFrame(df_grouping.sort_values("Diff.", ascending = False).head(1))
distrito = resp.index.values[0]
print(f"El Distrito {distrito} tiene una mayor diferencia de peatones con un total de {resp.iloc[0]["Diff."]} peatones.")
```

```python
sns.violinplot(data=df_peatones, y="PEATONES", x="DISTRITO")
fig=plt.gcf()
fig.set_size_inches(14, 7)
print("El Distrito Centro tiene mayor difrencia de peatones.")
```

```python
df_grouping2 = df_peatones[df_peatones["DISTRITO"] == "Chamberí"].groupby(["DISTRITO", "HORA"])["PEATONES"].agg("mean")
```
