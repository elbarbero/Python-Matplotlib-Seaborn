```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df_temp = pd.read_excel(r'C:\Users\ivan_pinar\Dropbox\Creación de MOCs\MOC Master Python Análisis de Datos\Datasets\8.7\Temperaturas.xlsx',sheet_name="Sheet1")
```

```python
df_temp["T.Promedio"]=(df_temp["T. Máxima"]+df_temp["T.Mínima"])/2
```

```python
df_temp.head()
```

```python
df_temp=df_temp[df_temp["Ciudad"]=="Madrid"]
```

```python
window_size = 10
```

```python
windows = df_temp["T.Promedio"].rolling(window_size)
```

```python
moving_averages = windows.mean()
```

```python
moving_averages_list = moving_averages.tolist()
```

```python
moving_averages_list
```

```python
plt.plot(df_temp["FECHA"],df_temp["T.Promedio"],label='T.Pro_Día')
plt.plot(df_temp["FECHA"],moving_averages_list,label='Media Móvil Temperatura n=10')
plt.legend()
```
